# 3. Basic Knowledge_Graphics Python

## matplotlib library

```python
import matplotlib.pyplot as plt
```

```python
# generate data
import numpy as np

x=np.random.randn(100)
y=np.random.randn(100)
```

## 3.1. plot

```python
fig, ax= plt.subplots(figsize=(8,8))
ax.plot(x,y)
```

## 3.2. Scatterplot

### 3.2.1 Method1_ ax.plot(x,y, 'x')

```python
fig, ax= plt.subplots(figsize=(8,8))
ax.plot(x,y, 'o')
```

### 3.2.2. method 2_ *ax.scatter(x,y, marker='x')*

```python
fig, ax= plt.subplots(figsize=(8,8))
ax.scatter(x,y, marker='x')
```

## 3.3. set label and title


```python
fig, ax= plt.subplots(figsize=(8,8))
ax.plot(x,y, marker='o')

ax.set_xlabel('X_Axis')
ax.set_ylabel('Y_Axis')
ax.set_title("plot of X vs Y")
```

## 3.4. change fig size

```python
fig.set_size_inches(12,3)
fig
```

## 3.5. Subplots

```python
fig, axes=plt.subplots(nrows=2, ncols=3, figsize=(15,5))
```

```python
axes[0,0].plot(x,y,'o', c='r')
axes[0,1].plot(x,y,'x', c='b')
axes[0,2].plot(x,y,'+', c='g')
axes[1,0].plot(x,y,marker='o', c='r')
axes[1,1].plot(x,y,marker='x', c='b')
axes[1,2].plot(x,y,marker='+', c='g')

fig
```

 ## 3.6. Save figure

### 3.6.1 Save as image

```python
fig.savefig("Figure_exmp_1.png", dpi=200)
fig.savefig("Figure_exmp_1.jpg", dpi=200)
```

### 3.6.2 Save as pdf

 ## 3.7. Determine max and min of axes

```python
axes[0,1].set_xlim([-1,1])
axes[0,1].set_ylim([-2,2])
fig
```

## 3.8. Countour Plot

```python
x=np.linspace(-np.pi, np.pi, 50)
y=x
f=np.multiply.outer(np.cos(y),1/(1+x**2))
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.contourf(x,y,f)
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.contour(x,y,f)
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.contour(x,y,f, levels=45)
```
